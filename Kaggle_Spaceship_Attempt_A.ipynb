# The Model

```python
input=tf.keras.layers.Input(20)
#hidden1=tf.keras.layers.Dense(2,activation='relu')(input)
output=tf.keras.layers.Dense(1,activation='sigmoid')(input)
```

```python
model=tf.keras.Model(inputs=input,outputs=output)
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.001), loss=tf.compat.v1.losses.log_loss,metrics=['accuracy'])
```

```python
model.fit(df_final,y_final,epochs=20)
```
